## Experiment 7 :
Write a Pandas program to get the powers of an array values element-wise. 

```python
import pandas as pd
```

```python
df=pd.DataFrame({'X':[2,3,4,2,10], 'Y':[2,3,3,4,2],'Z':[2,3,2,4,3]})
df
```

```python
p = (df.X**df.Y)**df.Z
```

```python
print(p)
```

```python
df['Power_ans'] = p
df
```
