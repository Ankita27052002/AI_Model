```python
df.horsepower.plot(kind = 'hist',bins = 3)
```

```python
count , bins = np.histogram(df.horsepower, 3)
```

```python
label = ['low','medium','high']
```

```python
df['horsepower-binned'] = pd.cut(df.horsepower, bins, labels = label)
```
